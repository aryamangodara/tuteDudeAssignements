# Importing data and creating the data frames

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
movies.head()
# credits.head()
```

```python
data = movies.merge(credits, on='title')
data.head()
data = data[[ 'title','release_date', 'genres','id', 'keywords', 'overview', 'cast', 'crew' ]]
data.dropna(inplace=True)
data.isnull().sum()
data.info()
data.sample(5)
```

```python
def extract_members(col):
    members = []
    for item in eval(col):
        members.append(item['name'])
    return members
```

```python
data['genres'] = data['genres'].apply(extract_members)
data['keywords'] = data['keywords'].apply(extract_members)
```

```python
data.head(5)
```

```python
def top3Cast(col):
    cast = []
    count = 1
    for item in eval(col):
        cast.append(item['name'])
        count += 1
        if count > 3:
            break
    return cast

data['cast'] = data['cast'].apply(top3Cast)
data.head(5)
```

```python
def extract_director(col):
    for item in eval(col):
        if item['department'] == 'Directing':
            return item['name']
    return np.nan

data['director'] = data['crew'].apply(extract_director)
data.sample(5)
```

```python
data.drop(['crew'], axis=1, inplace=True)
data.head(5)
```

```python
data['overview'] = data['overview'].apply(lambda x: x.split(' '))
data.head(5)
```

```python
# data['director']= data['director'].apply(lambda x: x.replace(' ',''))
data.cast[0]
data['cast']=data['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
data.head(5)
data['director'] = data['director'].fillna('')
data['director'] = data['director'].apply(lambda x: [x.replace(" ", "")])
data['keywords'] = data['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
data.head(5)
```

```python
data['genres'] = data['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
data.sample(5)
```

```python
data['tags']= data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['director']
data.head(5)
```

```python
data['tags'][0]
```

```python
data.drop(['overview', 'genres', 'keywords', 'cast', 'director','release_date'], axis=1, inplace=True)
data.head(5)
```

```python
" ".join(data['tags'][0]).lower()
data['tags']=data['tags'].apply(lambda x: " ".join(x).lower())
data.head()
data['tags'][0]
```

```python
data['title']=data['title'].apply(lambda x: x.lower())
data.head()
```

# NLP -> nltk library

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemming(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
data['tags'] = data['tags'].apply(stemming)
```

```python
data['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(data['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()[:100]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = data[data['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(data.iloc[i[0]].title)
recommend('meet dave')
```

```python
recommend('the dark knight rises')
```
