```python
import matplotlib.pyplot as plt
x = [1,2,3,4]

y1 = [10,20,25,30]
y2 = [5,15,20,25]
y = [5,15,20,25]

plt.plot(x,y1,label="Product A")
plt.plot(x,y2,label="Product B")

plt.legend()

plt.show()
plt.scatter(x,y1,label="Product A")
plt.show()
```

```python
categories = ["A","B","C"]
values = [10,20,15]

plt.bar(categories,values)

plt.show()
```

```python
sizes = [40,30,20,10]
labels = ["A","B","C","D"]

plt.pie(sizes,labels=labels,autopct="%1.1f%%")

plt.show()
```

```python
import numpy as np

x = np.linspace(0,10,100)

y = np.sin(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

```python
fig , axis = plt.subplots(2,2,figsize=(10,8))

axis[0,0].plot(x,y)
axis[0,0].set_title("Sine Wave")
axis[0,0].set_xlabel("X-axis")
axis[0,0].set_ylabel("Y-axis")
plt.grid()
axis[0,1].bar(categories,values)
axis[0,1].set_title("Bar Chart")
axis[1,0].hist(data,bins=30)
axis[1,0].set_title("Histogram")
axis[1,0].grid()
axis[1,0].set_xlabel("Value")
axis[1,0].set_ylabel("Frequency")
axis[1,1].pie(sizes,labels=labels,autopct="%1.1f%%")
axis[1,1].set_title("Pie Chart")
plt.tight_layout()
plt.show()
```
