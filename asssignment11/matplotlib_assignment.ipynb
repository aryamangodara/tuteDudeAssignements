# Task 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bmw_global_sales_2018_2025.csv')
print(df.sample(10))  
```

```python
plot_no = 1
def show_fig():
    plt.tight_layout()
    plt.show()
    global plot_no
    plot_no += 1
```

```python
import calendar
df_x5_usa_2024 = df[(df['Year'] == 2024) & (df['Model'] == 'X5') & (df['Region'] == 'USA')]
plt.plot(df_x5_usa_2024['Month'], df_x5_usa_2024['Units_Sold'],marker='o', linestyle='-')
plt.xlabel('Month')
month_abbrs = [calendar.month_abbr[i] for i in range(1, 13)]
plt.xticks(ticks=range(1, 13), labels=month_abbrs)
plt.ylabel('Units Sold')
plt.title(f'{plot_no}. BMW Monthly Sales Of X5 in USA (2024)')
show_fig()
```

# Task 2

```python
china_df = df[(df['Region'] == 'China')]
plt.scatter(china_df['Fuel_Price_Index'],china_df['BEV_Share'])
plt.legend()
plt.xlabel('Fuel Price Index')
plt.ylabel('BEV Share')
plt.title(f'{plot_no}. Fuel Price vs. BEV Share in China')
show_fig()
```

# Task 3

```python
df_model_sales = df.groupby('Model')['Units_Sold'].sum().reset_index()
plt.bar(df_model_sales['Model'], df_model_sales['Units_Sold'].sort_values(ascending=False), color='skyblue')
plt.xlabel('Model')
plt.ylabel('Units Sold')
plt.title(f'{plot_no}. BMW Model-wise Sales')
plt.grid(axis='y')
show_fig()
```

```python
plt.barh(df['Model'], df['Units_Sold'])
plt.xlabel('Model')
plt.ylabel('Units Sold')
plt.title(f'{plot_no}. BMW Model-wise Sales')
show_fig()
```

# Task 4

```python

df_2024 = df[df['Year'] == 2024]
df_2025 = df[df['Year'] == 2025]

sales_2024 = df_2024.groupby('Model')['Units_Sold'].sum().reset_index()
sales_2025 = df_2025.groupby('Model')['Units_Sold'].sum().reset_index()

x = np.arange(len(sales_2024['Model']))
width = 0.35

plt.bar(x - width/2, sales_2024['Units_Sold'], width=width, label='2024', color='skyblue')
plt.bar(x + width/2, sales_2025['Units_Sold'], width=width, label='2025', color='lightcoral')

plt.xticks(ticks=x, labels=sales_2024['Model'], rotation=45) 


plt.xlabel('Model')
plt.ylabel('Total Units Sold') 
plt.title(f'{plot_no}. BMW Model-wise Sales')
plt.legend()
plt.grid(axis='y')
plt.figure(figsize=(30, 30))
show_fig()
```

# Task 5

```python
plt.bar(sales_2024['Model'], sales_2024['Units_Sold'],label='2024',color='skyblue')
plt.bar(sales_2025['Model'], sales_2025['Units_Sold'],bottom=sales_2024['Units_Sold'],color='red',label='2025')
plt.xlabel('Model')
plt.ylabel('Units Sold')
plt.title(f'{plot_no}. Stacked BMW Model-wise Sales')
show_fig()
```

# Task 6

```python
plt.hist(df['Avg_Price_EUR'], bins=30, color='skyblue', edgecolor='black', label='Average Price Distribution')
plt.xlabel('Average Price (EUR)')
plt.ylabel('Frequency')
plt.title(f'{plot_no}. BMW Average Price Distribution')
plt.legend()
show_fig()
```

```python
df_yearly_sales = df.groupby(['Year'])['Units_Sold'].sum().reset_index()
plt.pie(df_yearly_sales['Units_Sold'], labels=df_yearly_sales['Year'], autopct='%1.1f%%', startangle=140)
plt.title(f'{plot_no}. BMW Yearly Sales Distribution')
show_fig()
```
