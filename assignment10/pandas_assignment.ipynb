# Task 1

```python
import pandas as pd
marks = [75,80,95,66,72]
pd_series = pd.Series(marks)
print(pd_series.values)
print(pd_series.index)
print(pd_series.dtype)
print(pd_series[0])
print(pd_series.tail(2))
```

# Task 2


```python
print(pd_series+5)
print(pd_series-2)
print(pd_series*1.05)
print(pd_series/2)
```

# Task 3

```python
print(pd_series.max())
print(pd_series.min())
print(pd_series.sum())
print(pd_series.mean())
```

```python
result = pd_series.apply(lambda x: x > 70)
print(result)
print(result.sum())
```

# Task 4

```python
students={
    'Name' : ['Amit', 'Neha', 'Rahul', 'Sneha', 'Pooja'],
    'Marks' : [75,80,95,66,72],
    'Subject' : ['Math', 'Math', 'Science', 'Math', 'Math']
}
df = pd.DataFrame(students)
print(df)
```

```python
print(df.head(3))
print(df.tail(2))
```

```python
print(df.shape)
print(df.columns.values)
```

# Task 5

```python
print(df.info())
print(df.describe())
print(df.head())
print(df.tail())
```

```python
df.sort_values(by='Marks', ascending=False,inplace=True)
print(df)
print(df.reset_index(drop=True))
```

# Task 6

```python
print(df[df['Marks'] > 75])
print(df[df['Subject'] == 'Math'])
print(df[df['Marks'] > df['Marks'].mean()])
print(df[df['Marks']<70])
```

# Task 7

```python
print(df.groupby('Subject')['Marks'].mean())
print(df.groupby('Subject')['Name'].count())
print(df.groupby('Subject')['Marks'].max())
```

# Task 8

```python
df.plot(title='Student Marks', x='Name', y='Marks', kind='bar')
```

# Task 9

```python
sales = {
    'Day':['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'Revenue':[1200, 1500, 900, 2000, 1800]
}
sales_df = pd.DataFrame(sales)
print(sales_df['Revenue'].sum())
print(sales_df['Revenue'].mean())
print(sales_df[sales_df['Revenue']==sales_df['Revenue'].max()]['Day'])
print(sales_df[sales_df['Revenue']>sales_df['Revenue'].mean()]['Day'])
sales_df.plot(title='Daily Revenue', x='Day', y='Revenue')
```
