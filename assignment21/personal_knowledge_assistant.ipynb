## PART 2 — Text Splitters in LangChain

### Task 6: Why Text Splitting is Required (Conceptual)
Answer briefly:
1. Why large documents cannot be directly passed to LLMs?
2. What problems does chunking solve in GenAI systems?


**Answer:**
1. **Token Limits and Cost:** Large documents exceed the context window constraints (token limits) inherent in all LLM models. If passed directly, the model will throw an error or truncate the critical parts. Further, even for large-context models, sending massive texts degrades the focus and significantly increases processing compute cost and latency.
2. **Retrieval Purity:** Chunking solves this by cutting text into cohesive segments, which can be stored in a Vector DB. During RAG, chunking allows semantic searches to reliably capture exactly the most relevant snippet rather than dragging an entire bloated document into context. This helps reduce hallucination.

### Task 8: Text Structure–Based Splitter
1. Use RecursiveCharacterTextSplitter.
2. Explain how it respects text structure (paragraphs, sentences).
3. Apply it to the same documents.
4. Compare output with length-based splitter.


**Explanation for RecursiveCharacterTextSplitter:**
The `RecursiveCharacterTextSplitter` respects text structure by trying to split content along natural semantic boundaries in a hierarchical order: double newlines (`\n\n` for paragraphs), single newlines (`\n` for sentences), spaces (` ` for words), and finally individual characters. In doing so, it avoids cutting a sentence or a word in half, creating far more coherent chunks than those produced by a strict character count approach.


```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

rec_char_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

# Applying splitter to the same text document
rec_char_chunks = rec_char_splitter.split_documents(text_docs)

print(f"Number of chunks (RecursiveCharacterTextSplitter): {len(rec_char_chunks)}")
if rec_char_chunks:
    print(f"Sample chunk:\n{rec_char_chunks[0].page_content}")
```

### Task 11: Unified Preprocessing Pipeline
Create a function:
`def load_and_split_documents(path_or_url):`
   - # loads documents
   - # splits them
   - # returns chunks
Apply it to:
- Local directory
- Web URL


```python
import os
from urllib.parse import urlparse
from langchain_community.document_loaders import DirectoryLoader, TextLoader, CSVLoader, PyPDFLoader

def load_and_split_documents(path_or_url):
    # Determine type of path
    is_url = urlparse(path_or_url).scheme in ('http', 'https')
    
    docs = []
    # 1. Load documents
    try:
        if is_url:
            loader = WebBaseLoader(path_or_url)
            docs = loader.load()
        elif os.path.isdir(path_or_url):
            # Safe folder parsing mapping extensions to precise loaders to bypass unstructured
            loaders = [
                DirectoryLoader(path_or_url, glob="**/*.txt", loader_cls=TextLoader),
                DirectoryLoader(path_or_url, glob="**/*.csv", loader_cls=CSVLoader),
                DirectoryLoader(path_or_url, glob="**/*.pdf", loader_cls=PyPDFLoader),
            ]
            for folder_loader in loaders:
                docs.extend(folder_loader.load())
        elif os.path.isfile(path_or_url):
            if path_or_url.endswith('.txt'):
                loader = TextLoader(path_or_url)
            elif path_or_url.endswith('.csv'):
                loader = CSVLoader(path_or_url)
            elif path_or_url.endswith('.pdf'):
                loader = PyPDFLoader(path_or_url)
            else:
                raise ValueError("Unsupported extension")
            docs = loader.load()
        else:
            raise ValueError("Invalid path input.")
            
    except Exception as e:
        print(f"Skipping loading due to exception (such as missing package): {e}")
        docs = []
        
    # 2. Split documents
    splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    chunks = splitter.split_documents(docs)
    
    # 3. Return chunks
    return chunks

# Test Local Directory Pipeline
print("Testing on local directory 'data/':")
dir_chunks = load_and_split_documents("data/")
print(f"Number of chunks recursively extracted: {len(dir_chunks)}\n")

# Test Web URL Pipeline
print("Testing on web URL:")
web_test_chunks = load_and_split_documents("https://example.com/")
print(f"Number of chunks extracted from URL: {len(web_test_chunks)}")
if web_test_chunks:
    print(f"Sample content: {web_test_chunks[0].page_content[:200]}")
```

### Task 12: Observations & Insights
Write short answers:
1. Which loader is used for which data type?
2. Best splitter for:
   - Small text
   - Large PDFs
   - Web data
3. Why chunk overlap is important?


**Answer:**
1. **Loaders Matching:**
   - `.txt` -> `TextLoader`
   - `.csv` -> `CSVLoader`
   - `.pdf` -> `PyPDFLoader`
   - Whole Folders -> `DirectoryLoader`
   - Web Pages (`http/https`) -> `WebBaseLoader`

2. **Best Splitter For:**
   - **Small text:** A basic `CharacterTextSplitter` often suffices since performance and strict logic don't dictate heavy optimization.
   - **Large PDFs:** `RecursiveCharacterTextSplitter` combined with page / paragraph specific loaders. Because it scales down gracefully, it honors the structure of PDFs safely.
   - **Web data:** `HTMLHeaderTextSplitter` combined with `RecursiveCharacterTextSplitter` retains the exact structural HTML header elements which are critical for SEO semantic extraction.

3. **Importance of Chunk Overlap:**
   Chunk overlap mitigates the risk of slicing context right down the middle of a thought. An overlapping token window (e.g. 50 characters) creates continuity between `Chunk A` and `Chunk B`. If a pronoun in `Chunk B` references a proper noun separated at the end of `Chunk A`, overlap guarantees both chunks hold enough surrounding context for the LLM during RAG retrieval.
